**Exercise 1**

*1.1* (5 pts) Read in the 'county_age_dist.csv' and cast the 'fips' column as a string such that it is a length 5 string of digits that starts with 0 if it is currently of length 4.  Yolo county for example is '06113'.  This is the age distribution of each county in the US (with territories), you may notice that the total population is unusually large; please ignore this.

```python
import pandas as pd
import numpy as np
```

```python
age_dist = pd.read_csv('county_age_dist.csv',dtype={'fips':str})
age_dist['fips'] =  age_dist['fips'].apply(lambda x : '{0:0>5}'.format(x))
age_dist = age_dist.set_index(['fips'])
age_dist
```

*1.2* (5 pts) Create a population column which is the sum of the age counts for all age bins.  For each age bin create a column which is proportion of the age bin in the county populations, so for example '0-17_rat' is the ratio of '0-17' to the whole population for the county.  Print out the FIPS code for the county with the largest proportion of 0-17 year olds. What is the FIPS code for the county with the largest proportion of 85+ year olds?

```python
age_dist['populations'] = age_dist.loc[:,:'85+'].sum(axis=1)
for i in range(9):
    age_dist[age_dist.columns[i]+'_rat'] = age_dist.iloc[:,i] / age_dist['populations'] 
age_dist
```

```python
print("FIPS " + age_dist['0-17_rat'].idxmax() + " has the largest proportion of 0-17 year olds.")
print("FIPS " + age_dist['85+_rat'].idxmax() + " has the largest proportion of 85+ year olds.")
```

*1.3* (10 pts) What is the overall proportion of each age bin within the entire US.  You need to find the total number of each age bin across all counties and take its ratio with the total population.

```python
tot_pop = age_dist['populations'].sum()
for i in range(9):
    print("Age bin " + age_dist.columns[i] + " overall proportion is " + str(age_dist.iloc[:,i].sum() / tot_pop))
```

*1.4* (15 pts) Create a new column that contains the mode of the age bins (the age bin with the largest frequency in the county).  Print out the frequencies of the modes' different possible values, so that the frequency of '0-17' is the proportion of the counties that have '0-17' as the largest age bin for that county.

```python
age_dist['mode'] = age_dist.iloc[:,:9].idxmax(axis = 1)
from collections import Counter
c = Counter(age_dist['mode'])
c_sum = sum(c.values())
for freq in c:
    print("Age bin "+freq+" frequency: "+str(c[freq]/c_sum))
```

*2.1* (5 pts) Read in the 'county_fips_master.csv' and join this to the county age dataframe constructed above such that there is one row for every county in the new dataframe from county_fips_master.  How many rows are there in the new dataframe?  Are there any missing entries?  If so how many rows have missingness?

```python
fips_master = pd.read_csv('county_fips_master.csv',converters={'fips':str})
fips_master['fips'] =  fips_master['fips'].apply(lambda x : '{0:0>5}'.format(x))
fips_master = fips_master.set_index(['fips'])
join_df = fips_master.join(age_dist,how='left')
print("There are " + str(len(join_df)) + " rows in the new dataframe.")
print("Are there any missing entries? " + str(join_df.isnull().values.any()))
print(str(join_df.isnull().any(axis=1).sum()) + " rows have missingness.")
```

*2.2* (10 pts) Drop any rows with missingness.  For each state calculate the proportion of each age bin as you did for the entire US in the question above.  What is the state with the most 0-17 proportion and the state with the most 85+ proportion?  Give the proportions.

```python
join_df_drop = join_df.dropna(axis=0)
print(join_df_drop.shape)
groupby_sum = join_df_drop.groupby(["state_name"]).sum()
print("The state with the most 0-17 proportion is " + (groupby_sum['0-17']/groupby_sum['populations']).idxmax())
print("The state with the most 85+ proportion is " + (groupby_sum['85+']/groupby_sum['populations']).idxmax())
```

*3.1* (10 pts) Read in 'time_series_covid19_deaths_US.csv'.  This is the Johns Hopkins official US Covid-19 death cumulative counts for each date and county.  Cumulative cases mean the total reported cases up to and including that date.  Restructure the dataframe such that it is 'tidy' and there is a column for date, deaths, population, and fips.  (Hint: look at pandas melt function)

```python
covid19_deaths = pd.read_csv('time_series_covid19_deaths_US.csv')
covid19_deaths = covid19_deaths.dropna(subset=['FIPS'])
covid19_deaths_tidy = pd.melt(covid19_deaths, id_vars=['FIPS','Population'], value_vars=covid19_deaths.columns[12:], \
                             var_name= 'date', value_name='deaths')
covid19_deaths_tidy['FIPS'] = covid19_deaths_tidy['FIPS'].astype(np.int64)
```

```python
covid19_deaths_tidy['FIPS'] =  covid19_deaths_tidy['FIPS'].apply(lambda x : '{0:0>5}'.format(x))
covid19_deaths_tidy = covid19_deaths_tidy.set_index(['FIPS'])
covid19_deaths_tidy.head()
```

```python
covid19_deaths_tidy.shape
```

*3.2* (5 pts) Join the death data to the dataset with the county names, such that it drops those counties that are not within the county name dataset (county_fips_master).  Use only these counties henceforth.  This should remove any counties that have 0 population.  Which county (give county name and state) has the highest proportion of deaths by 2/25/21 to the Population?  Use the population in the covid deaths datasets.

```python
join_df_2 = fips_master.join(covid19_deaths_tidy,how='left')

max_2_25_21_fips = (join_df_2.loc[join_df_2['date'] == '2/25/21']['deaths'] \
    / join_df_2.loc[join_df_2['date'] == '2/25/21']['Population']).idxmax()

print("The county has the highest proportion of deaths by 2/25/21 to the Population: ")
join_df_2.loc[join_df_2.index.isin([max_2_25_21_fips]) & (join_df_2.date == '2/25/21')][['county_name','state_name']]
```

*3.3* (10 pts) Compute the incidence proportion (number of deaths each day per 100,000 population) for each county, you may save this as a new dataset.  You can calculate the incidence for a given county by taking the difference of the cumulative cases for a day with the previous day.  Notice that this dataset is quite large, one trick to make this faster is to use the fips code and date as indexing variables and use `sort_index` to sort the indices before applying groupby operations.  For full points, you should use `groupby` and `diff` for this operation.

```python
date_time = pd.to_datetime(join_df_2['date']) 
join_df_2['date'] = date_time
```

```python
join_df_2 = join_df_2.reset_index().rename(columns={'index': 'FIPS'}).set_index(['FIPS','date'])
join_df_2 = join_df_2.sort_index()
join_df_2.head()
```

```python
incid_prop = join_df_2.eval('deaths / Population * 100000').groupby(level = 0).diff()
incid_prop = incid_prop.to_frame().rename({0: 'incidence_proportion'}, axis=1)
```

```python
incid_prop.head()
```

*3.4* (15 pts) Compute the incidence proportion per week (per 100,000 population), which is the sum of the incidences for all days in a week.  Hint: use Periods in Pandas.  A week is defined to be Sunday through Saturday.  Which is the week with the highest mean incidence proportions (where the mean is taken over all the counties)?  You can either sum the incidence counts within the week, or take a difference of the cumulative cases between consecutive Sundays, and these should produce the same thing.

```python
incid_prop = incid_prop.reset_index(level=1)
incid_prop['week']=pd.PeriodIndex(incid_prop['date'], freq='W')
```

```python
incid_prop.head()
```

```python
incid_prop_week = incid_prop.groupby(['week'])['incidence_proportion'].mean()
incid_prop_week.to_frame().head()
```

```python
incid_prop_week.to_frame().describe()
```

```python
print(str(incid_prop_week.to_frame()['incidence_proportion'].idxmax()) + \
      " is the week with the highest mean incidence proportions.")
```
